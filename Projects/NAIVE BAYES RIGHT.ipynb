```python
from sklearn.impute import SimpleImputer
si=SimpleImputer()
df['MinTemp']=si.fit_transform(df[['MinTemp']])
df['MaxTemp']=si.fit_transform(df[['MaxTemp']])
df['Rainfall']=si.fit_transform(df[['Rainfall']])
df['Evaporation']=si.fit_transform(df[['Evaporation']])
df['Sunshine']=si.fit_transform(df[['Sunshine']])
df['WindGustSpeed']=si.fit_transform(df[['WindGustSpeed']])
df['WindSpeed9am']=si.fit_transform(df[['WindSpeed9am']])
df['WindSpeed3pm']=si.fit_transform(df[['WindSpeed3pm']])
df['Humidity9am']=si.fit_transform(df[['Humidity9am']])
df['Humidity3pm']=si.fit_transform(df[['Humidity3pm']])
df['Pressure9am']=si.fit_transform(df[['Pressure9am']])
df['Pressure3pm']=si.fit_transform(df[['Pressure3pm']])
df['Cloud9am']=si.fit_transform(df[['Cloud9am']])
df['Cloud3pm']=si.fit_transform(df[['Cloud3pm']])
df['Temp9am']=si.fit_transform(df[['Temp9am']])
df['Temp3pm']=si.fit_transform(df[['Temp3pm']])
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
se=SimpleImputer(strategy='most_frequent')
df['WindGustDir']=se.fit_transform(df[['WindGustDir']]).ravel()
df['WindDir9am']=se.fit_transform(df[['WindDir9am']]).ravel()
df['WindDir3pm']=se.fit_transform(df[['WindDir3pm']]).ravel()
df['RainToday']=se.fit_transform(df[['RainToday']]).ravel()
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Location']=le.fit_transform(df['Location'])
df['WindGustDir']=le.fit_transform(df['WindGustDir'])
df['WindDir9am']=le.fit_transform(df['WindDir9am'])
df['WindDir3pm']=le.fit_transform(df['WindDir3pm'])
df['RainToday']=le.fit_transform(df['RainToday'])
df.info()
```

```python
x=df.drop(columns='RainToday',axis=1)
y=df['RainToday']
x
```

```python
from sklearn.preprocessing import MinMaxScaler
st=MinMaxScaler()
x=st.fit_transform(x)
x
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
