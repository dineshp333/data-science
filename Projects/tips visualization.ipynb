**PLOTS**

*FOR UNIVARIATE*

```python
plt.figure(figsize=(5,6))
sns.histplot(df['total_bill'],color='darkgreen',kde=True)
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.title('total bill amount')
plt.show()
```

my analysis in this plot:
This is a histogram .It is used to analysis continous data and it is a univariate diagram.people 10-30 most ah rupees bill pay pannirukanga and .It is normal distribution.

```python
plt.figure(figsize=(5,6))
sns.boxplot(data=df,x='tip',color='pink')
plt.title('tips')
plt.xlabel('tips')
plt.ylabel('frequency')
plt.show()
```

It is countplot .It is very usefull for detecting outliers in linear regression .people average ah 1 to 6 rupees kulla than tips tharanga .6 ku mela rare ah tips tharanga .so that is outliers

```python
pie=df['sex'].value_counts()
plt.figure(figsize=(5,6))
plt.pie(pie,labels=['male','female'],autopct='%1.1f%%',colors=['blue','red'],startangle=180,)
plt.title('gender')
plt.show()
```

This is pie chart.In this chart only for discrete data and it male customers gives more tips than female customers.

```python
sns.countplot(data=df,x='day',color='orange')
plt.title('Customer visiting days')
plt.xlabel('day')
plt.show()
```

This is countplot.It is best for categorical data and SMOTE method for imbalanced dataset.this plot shows people giving tips or visiting hotel frequently in saturday.very less visiting day is friday 

*for bivariate*

```python
sns.barplot(data=df,x='day',y='tip',color='violet')
```

This is bar plot.It is very usefull for comparing 2 features.Comparitively sunday most of the customers given tips.

```python
plt.figure(figsize=(5,6))
sns.scatterplot(data=df,x='total_bill',y='tip',style='sex',palette='coolwarm')
plt.title('relationship between total bill and tip')
plt.show()
```

*for multivariate*

```python
sns.pairplot(df[['total_bill','tip','sex','smoker']],hue='sex',palette='Set1',)
plt.suptitle('pair plot of titanic variables',y=1.02)
plt.show()
```

It is pairplot .It is allows more than 2 features for understanding in one plot.total bill male and female shows normal distribution.And tip shows these 2 are highly correlated.

```python
sns.violinplot(x='time',y='tip',hue='sex',data=df,split=True,palette='Set3')
plt.title('comparing time and gender with age')
plt.legend(title='Gender')
plt.show()
```

This is Violin plot .It is helps to comparing more than 2 features .My analysis in this output is lunch time tips providers is greater than the dinner ptips providers.And mostly male customers gives more tips comparitively female customers.Both of lunch and dinner time 2-3 rupees tips given by customers.In dinner time only customer gives tips more than 6 rupees.
