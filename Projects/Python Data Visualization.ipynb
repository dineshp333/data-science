```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'D:\Data Science With Python\dataset\Linear dataset\Smartphones_cleaned_dataset.csv')
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(5,6))
sns.histplot(df['screen_size'],kde=True,color='red')
plt.title('Histoggram for rating')
plt.show()
```

```python
sns.boxplot(x='rating',data=df,color='green')
plt.ylabel('count')
plt.title('Ratings from customers')
plt.show()
```

```python
df['os'].unique()
```

```python
pie=df['os'].value_counts()
plt.pie(pie,labels=['android','ios','other'],autopct='%1.1f%%',colors=['red','blue','yellow'],startangle=180)
plt.title('Pie diagram for OS')
plt.show()
```

```python
sns.countplot(data=df,x='os',color='orange')
plt.title('Count of the os')
plt.show()
```

```python
sns.barplot(data=df,x='os',y='price',color='darkblue')
plt.title('Price differences in OS')
plt.show()
```

```python
sns.kdeplot(df['num_cores'],shade=True,color='red')
plt.title('Number of cores density')
plt.show()
```

```python
plt.figure(figsize=(4,9))
sns.pairplot(df[['price','rating','processor_brand','os']],hue='os',palette='Set1')
plt.title('Comparing rating and price with os',y=2.1)
plt.show()
```
