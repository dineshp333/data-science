Data Visualization with Python

*import libraries*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

*import dataset*

```python
df=pd.read_csv(r'D:\Data Science With Python\dataset\Linear dataset\Smartphones_cleaned_dataset.csv')
df.head()
```

```python
df.shape
```

Different types of plots

```python
plt.figure(figsize=(5,6))
sns.histplot(df['screen_size'],kde=True,color='red')
plt.title('Histoggram for rating')
plt.show()
```


This is histogram.It is used to analysis continous data and it is a univariate diagram.It shows most of the mobile screen size in between range og 6-7inch.And it shows it is a normal distribution

```python
sns.boxplot(x='rating',data=df,color='green')
plt.ylabel('count')
plt.title('Ratings from customers')
plt.show()
```

This is boxplot.It is very usefull for detecting outliers.This graph shows most of the ratings in between range of 75-80


```python
df['os'].unique()
```

```python
pie=df['os'].value_counts()
plt.pie(pie,labels=['android','ios','other'],autopct='%1.1f%%',colors=['red','blue','yellow'],startangle=180)
plt.title('Pie diagram for OS')
plt.show()
```

This is pie chart.It is very helpfull for Categorical data.This graph shows 94% people buy android mobile and rest of the percentage prefered ios and other.

```python
sns.countplot(data=df,x='os',color='skyblue')
plt.title('Count of the os')
plt.show()
```

This is countplot.It is best for categorical data and SMOTE method for imbalanced dataset.This graph also shows most of the people bought android mobile.

```python
sns.barplot(data=df,x='os',y='price',color='darkblue')
plt.title('Price differences in OS')
plt.show()
```

This is Barplot.It is just like countplot.But is helps to findout relatinshiop between 2 features.This graph shows ios more costly than android and other.

```python
sns.kdeplot(df['num_cores'],shade=True,color='red')
plt.title('Number of cores density')
plt.show()
```

This is kdeplot.It is helps for detection distribution of the feature .This Chart Shows me this column is 'Normal Distribution'

```python
plt.figure(figsize=(4,9))
sns.pairplot(df[['price','rating','processor_brand','os']],hue='os',palette='Set1')
plt.title('Comparing rating and price with os',y=2.1)
plt.show()
```

This is pairplot and it is for multivariate.This graph shows most phones are priced low, but iOS models tend to have higher ratings and slightly higher prices.Android dominates the dataset but shows a wider spread in both price and rating.

```python
sns.violinplot(data=df,x='has_5g',y='rating',palette='Set1',split=True,hue='has_nfc')
plt.title('Distribution of Ratings by 5G and NFC Availability')
plt.show()
```

This is violinplot.It is helps to find out a distribution of 2 or more features.This graph shows 5g mobiles got more ratings in between 80-90 but most of the mobile 5g not include nfc.

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x='rating',y='price',data=df,hue='os',style='has_5g',palette='Set1')
plt.title('Comparing rating and price with os')
plt.show()
```

A scatter plot displays individual data points on a graph to reveal relationships between two numerical variables.This graph shows ios devices costly comparitively android and other.Only 2 product is above 450000.

```python
sns.swarmplot(x='os',y='rating',data=df,hue='has_5g',palette='Set1')
plt.title('Ratings of the mobile based on OS')
plt.show()
```

A swarm plot displays individual data points within categories by spreading them out to show their exact values and density without overlap.This graph shows most of the mobiles got 75 and above ratings is 5g mobiles .And also ios mobiles almost has 5g .
